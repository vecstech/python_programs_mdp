````python
"""
Student grading systems
1. if marks in between 80 to 100 then A grade
2. if marks in between 60 to 80 then B grade
3. if marks in between 40 t0 60 then C grade
4. if marks in between 25 to 40 then D grade
5. if marks in between 0 to 25 then Fail
6. other than above marks output shoud be please enter valid marks
"""
marks=int(input("please enter marks here:"))
if (marks>=80 and marks<=100):
    print("A grade")
elif (marks>=60 and marks<80):
    print("B grade")
elif (marks>=40 and marks<60):
    print("C grade")
elif(marks>=25 and marks<40):
    print("D grade")
elif(marks>=0 and marks<25):
    print("Fail")
else:
    print("please enter valid marks")
````
