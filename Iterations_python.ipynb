# itearations or loops in python
1. iterations use to work with repeatative performances or repeatative tasks
2. itearations are two types
3. `for` loop
4. `while` loop

```python - for loop syntax
for iter_variable in sequence:
    print("for is keyword to start the iterations")
    print("iter variable is type of variable to iterate data from sequence")
    print("in is keyword and operator to check sequence is present or not")
    print("sequence is collection like string,list,tuple,dict...etc")
```

``` python - while loop syntax
while (condition):
    print("while is keyword to start the iterattions based on conditions")
    print("if conditions is True then start iterations")
```

````python
#write a program to print 1 to 100 numbers using for loop
for num in range(1,101):
    print(num,end=",")
````

````python
#write a program to find the vowels from string using for loop?
name=str(input("please enter your name here:"))
for char in name:
    if (char in "aeiouAEIOU"):
        print(char)
````

````python
#write a program to print non vowels from string using for loop?
name=str(input("please enter name here:"))
for c in name:
    if (c not in "aeiouAEIOU"):
        print(c)
````

````python
# write a program to find the even numbers in between 500 to 600 using for loop and seperated by ,?
for i in range(500,601):
    if (i%2==0):
        print(i,end=",")
````

````python
for j in range(500,601):
    if (j%2==1):
        print(j,end=",")
````
